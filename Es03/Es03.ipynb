Si richiede di calcolare tramite il Metodo Monte Carlo le opzioni Call e Put per una opzione Europea inizializzata coi seguenti parametri:

- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

Per calcolare le varie opzioni si usa un moto Browniano geometrico implementato con due metodi:
- metodo diretto
- metodo standard

Il metodo diretto prevede di calcolare direttamente il valore dell'opzione finale con la formula 
$$S(T) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$$ 
dove $W(T)$ Ã¨ una distribuzione normale centrata in $0$ e con varianza $T$.

Una volta ottenuto il valore finale si calcola il valore dell'opzione Call o Put col confronto con lo strike prize $K$ nel seguente modo:

se   $S(T)>K$    
$$C(T)=\exp[-r T] (S(T)-K)$$ 

$$P(T)=0$$
se, al contrario   $K>S(T)$      
$$C(T)=0$$   

$$P(T)=\exp[-r T] (K-S(T))$$

Dove con $C(T)$ si indicano le opzioni Call al tempo $T$ mentre con $P(T)$ le opzioni Put al tempo $T$

Tale procedimento viene fatto per $10^4$ volte e si mostrano i grafici ottenuti con il blocking method e relative incertezze (in azzurro) e il valore atteso (in rosso) calcolato con la soluzione analitica di Black-Scholes.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data_file=np.loadtxt('Data_Direct.dat')


x=data_file [:,0]
sum_progC=data_file [:,1]
err_progC=data_file[:,2]
sum_progP=data_file[:,3]
err_progP=data_file[:,4]


figure, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,10))
axes[0].errorbar(x,sum_progC,yerr=err_progC)
axes[0].set_xlabel('# experiment')
axes[0].set_ylabel('Call Options')
axes[0].set_title('Call Options, Direct Method')
axes[0].grid(True)
axes[0].axhline(y=14.975790778311286, color='r', linestyle='-')

axes[1].errorbar(x,sum_progP,yerr=err_progP)
axes[1].set_xlabel('# experiment')
axes[1].set_ylabel('Put Options')
axes[1].set_title('Put Options, Direct Method')
axes[1].grid(True)
axes[1].axhline(y=5.4595325819072364, color='r', linestyle='-')
plt.show()
```

Il metodo standard invece prevede di calcolare il valore dell'opzione ad ogni step temporale secondo la formula:

$$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$

dove $Z$ Ã¨ una distribuzione normale centrata in $0$ e con varianza $1$ ed ottenere cosÃ¬ il valore dell'opzione al tempo $T$ iterando il processo. 

Una volta ottenuto il valore finale dell'opzione si effettua il calcolo delle opzioni Call e Put come sopra.

Si effettua tale processo $10^4$ volte e si mostrano i grafici ottenuti con il blocking method e relative incertezze (in azzurro) e il valore atteso (in rosso) calcolato con la soluzione analitica di Black-Scholes.

```python
data_file_2=np.loadtxt('Data_Std.dat')
x2=data_file_2 [:,0]
sum_progC2=data_file_2 [:,1]
err_progC2=data_file_2 [:,2]
sum_progP2=data_file_2 [:,3]
err_progP2=data_file_2 [:,4]

figure, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,10))
axes[0].errorbar(x2,sum_progC2,yerr=err_progC2)
axes[0].set_xlabel('#throws')
axes[0].set_ylabel('Call Options')
axes[0].set_title('Call Options, Standard Method')
axes[0].grid(True)
axes[0].axhline(y=14.975790778311286, color='r', linestyle='-')

axes[1].errorbar(x2,sum_progP2,yerr=err_progP2)
axes[1].set_xlabel('#throws')
axes[1].set_ylabel('Put Options')
axes[1].set_title('Put Options, Standard Method')
axes[1].grid(True)
axes[1].axhline(y=5.4595325819072364, color='r', linestyle='-')
plt.show()
```
