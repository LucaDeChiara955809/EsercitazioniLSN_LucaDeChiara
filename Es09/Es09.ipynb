# Es 09.1

Si richiede di implementare un algoritmo genetico di modo da risolvere il problema del commesso viaggiatore.

Si richede di risolverlo in due casi:
- 32 città disposte su una circonferenza
- 32 città disposte all'interno di un quadrato

L'algoritmo è stato fatto con le seguenti caratteristiche:
- $N_{population}=200$
- $N_{generations}=10000$

Dove per $N_{population}$ si intende il numero di percorsi presi in esame per ogni generazione.

Per quanto riguarda gli algoritmi generici ne ho implementati 5, 4 di mutazione e uno di Crossing Over. Gli algoritmi di muazione prevedono:
- scambio di due città
- shift $m$ città contigue di $n$ posizioni
- inversione di ordine di m città contigue
- scambio di $m$ città contigue con altrettante

Si mostra dunque il miglior percorso trovato con l'algoritmo genetico e si mostra l'andamento della lunghezza del miglior percorso per ogni generazione e la media delle lunghezze dei migliori 50 percorsi per ogni generazione con errore calcolato a seguito di deviazione standard per i due casi.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

Niteration=10000
data_file=np.loadtxt('data_Circle.dat')

x=data_file [:,0]
y=data_file [:,1]

plt.errorbar(x,y)
plt.scatter(x,y,color='r')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Best Path")

plt.show() 
```

```python
data_file2=np.loadtxt('dataL_Circle.dat')

Lmin=data_file2 [:,0]
Lmean=data_file2 [:,1]
LdevS=data_file2[:,2]

x2 = np.arange(Niteration+1) 
figure, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,10))
axes[0].plot(Lmin)
axes[0].set_xlabel("$N_{generations}$")
axes[0].set_ylabel("$L_{min}$")
axes[0].set_title("$L_{min}$ Circle")

axes[1].errorbar(x2,Lmean,yerr=LdevS)
axes[1].set_xlabel("$N_{generations}$")
axes[1].set_ylabel("$L_{mean}$")
axes[1].set_title("$L_{mean}$ Circle")
plt.show() 
```

```python
data_file3=np.loadtxt('data_Square.dat')

x3=data_file3 [:,0]
y3=data_file3 [:,1]

plt.errorbar(x3,y3)
plt.scatter(x3,y3,color='r')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Best Path")
plt.show() 
```

```python
data_file4=np.loadtxt('dataL_Square.dat')

Lmin2=data_file4 [:,0]
Lmean2=data_file4 [:,1]
LdevS2=data_file4[:,2]

x2 = np.arange(Niteration+1) 
figure, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,10))
axes[0].plot(Lmin2)
axes[0].set_xlabel("$N_{generations}$")
axes[0].set_ylabel("$L_{min}$")
axes[0].set_title("$L_{min}$ Square")

axes[1].errorbar(x2,Lmean2,yerr=LdevS2)
axes[1].set_xlabel("$N_{generations}$")
axes[1].set_ylabel("$L_{mean}$")
axes[1].set_title("$L_{mean}$ Square")
plt.show() 
```

Si nota come le lunghezze dei percorsi diminuiscano rispetto alle prime generazioni ma non si assestano attorno a un minimo in quanto ho delle variazioni dovute alla casualità del processo.
