# Es 10.1

Si richiede di implementare un algoritmo di annealing di modo da risolvere il problema del commesso viaggiatore.

Si richede di risolverlo in due casi:
- 32 città disposte su una circonferenza
- 32 città disposte all'interno di un quadrato

L'algoritmo è stato fatto con le seguenti caratteristiche:
- $N_{population}=200$
- $N_{generations}=10000$

Dove per $N_{population}$ si intende il numero di percorsi presi in esame per ogni generazione.

Per quanto riguarda gli algoritmi generici ne ho implementati 5, 4 di mutazione e uno di Crossing Over. Gli algoritmi di muazione prevedono:
- scambio di due città
- shift $m$ città contigue di $n$ posizioni
- inversione di ordine di m città contigue
- scambio di $m$ città contigue con altrettante

Si mostra dunque il miglior percorso trovato con l'algoritmo genetico e si mostra l'andamento della lunghezza del miglior percorso per ogni generazione

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data_file=np.loadtxt('10.1/data_Circle.dat')
data_file2=np.loadtxt('10.1/dataL_Circle.dat')

x=data_file [:,0]
y=data_file [:,1]
Lmin=data_file2 [:]

figure, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,7.5))

axes[0].plot(x,y)
axes[0].scatter(x,y,color='r')
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].set_title("Best Path Circle")


axes[1].plot(Lmin)
axes[1].set_xlabel("$N_{iterations}$")
axes[1].set_ylabel("$L_{min}$")
axes[1].set_title("$L_{min}$ Circle")

plt.show()
```

```python
data_file3=np.loadtxt('10.1/data_Square.dat')
data_file4=np.loadtxt('10.1/dataL_Square.dat')

x2=data_file3 [:,0]
y2=data_file3 [:,1]
Lmin2=data_file4 [:]

figure, axes = plt.subplots(nrows=1, ncols=2,figsize=(15,7.5))

axes[0].plot(x2,y2)
axes[0].scatter(x2,y2,color='r')
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].set_title("Best Path Square")


axes[1].plot(Lmin2)
axes[1].set_xlabel("$N_{iterations}$")
axes[1].set_ylabel("$L_{min}$")
axes[1].set_title("$L_{min}$ Square")

plt.show()
```

Anche in questo caso, come nell'es 09.1, si nota come i valori dei percorsi diminuiscano rispetto alle prime generazioni ma non si assestano attorno ad un valore minimo ma ancora fluttuano, ciò è giustificato dalla casualità del processo. Si confrontano ora i risultati ottenuti con quelli dell'es 09.1:

```python
data9C=np.loadtxt('9_data_Circle.dat')
data9CL=np.loadtxt('9_dataL_Circle.dat')
data9S=np.loadtxt('9_data_Square.dat')
data9SL=np.loadtxt('9_dataL_Square.dat')

xc=data9C[:,0]
yc=data9C[:,1]
Lminc=data9CL[:,0]
xs=data9S[:,0]
ys=data9S[:,1]
Lmins=data9SL[:,0]

figure, axes = plt.subplots(nrows=4, ncols=2,figsize=(15,30))

axes[0,0].plot(xc,yc)
axes[0,0].scatter(xc,yc,color='r')
axes[0,0].set_xlabel("X")
axes[0,0].set_ylabel("Y")
axes[0,0].set_title("Best Path Circle Es 09")
axes[0,1].plot(x,y)
axes[0,1].scatter(x,y,color='r')
axes[0,1].set_xlabel("X")
axes[0,1].set_ylabel("Y")
axes[0,1].set_title("Best Path Circle Es 10")

axes[1,0].plot(Lminc)
axes[1,0].set_xlabel("$N_{iterations}$")
axes[1,0].set_ylabel("$L_{min}$")
axes[1,0].set_title("$L_{min}$ Circle es 09")
axes[1,1].plot(Lmin)
axes[1,1].set_xlabel("$N_{iterations}$")
axes[1,1].set_ylabel("$L_{min}$")
axes[1,1].set_title("$L_{min}$ Circle es 10")

axes[2,0].plot(xs,ys)
axes[2,0].scatter(xs,ys,color='r')
axes[2,0].set_xlabel("X")
axes[2,0].set_ylabel("Y")
axes[2,0].set_title("Best Path Square Es 09")
axes[2,1].plot(x2,y2)
axes[2,1].scatter(x2,y2,color='r')
axes[2,1].set_xlabel("X")
axes[2,1].set_ylabel("Y")
axes[2,1].set_title("Best Path Square Es 10")

axes[3,0].plot(Lmins)
axes[3,0].set_xlabel("$N_{iterations}$")
axes[3,0].set_ylabel("$L_{min}$")
axes[3,0].set_title("$L_{min}$ Square es 09")
axes[3,1].plot(Lmin2)
axes[3,1].set_xlabel("$N_{iterations}$")
axes[3,1].set_ylabel("$L_{min}$")
axes[3,1].set_title("$L_{min}$ Square es 10")

plt.show()
```

Si nota che i percorsi nel quadrato sono leggermente diversi, ciò è dovuto alla casualità della misura, ma nel complesso i risultati sembrano essere coerenti tra i due metodi.

# Es 10.2

Si richiede di implementare un algoritmo genetico di modo da risolvere il problema del commesso viaggiatore in programmazione parallela con 4 nodi.

Si richede di risolvere il problema per 32 città disposte all'interno di un quadrato.

L'algoritmo è stato fatto con le seguenti caratteristiche:
- $N_{population}=200$
- $N_{generations}=10000$

Dove per $N_{population}$ si intende il numero di percorsi presi in esame per ogni generazione.

Per quanto riguarda gli algoritmi generici ne ho implementati 5, 4 di mutazione e uno di Crossing Over. Gli algoritmi di muazione prevedono:
- scambio di due città
- shift $m$ città contigue di $n$ posizioni
- inversione di ordine di m città contigue
- scambio di $m$ città contigue con altrettante

Rispetto al problema $09.1$ sono stati presi diversi accorgimenti:
Per quanto riguarda la disposizione delle città, poichè generata casualmente, essa è stata generata soltanto per il nodo 0 e successivamente trasmessa agli altri nodi tramite una Broadcast, così da avere la stessa disposizione delle città. 

Inoltre era richiesto che dopo $N_{migr}$ bisognava scambiare i migliori elementi delle generazioni, per questo è stato necessario implementare due funzioni di send e recv di modo da poter trasmettere il messaggio tra i nodi.

Si mostra dunque il miglior percorso trovato con l'algoritmo genetico e si mostra l'andamento della lunghezza del miglior percorso per ogni generazione e la media delle lunghezze dei migliori 50 percorsi per ogni generazione con errore calcolato a seguito di deviazione standard.

```python
data0=np.loadtxt('10.2/data0.dat')
data1=np.loadtxt('10.2/data1.dat')
data2=np.loadtxt('10.2/data2.dat')
data3=np.loadtxt('10.2/data3.dat')

x0=data0 [:,0]
y0=data0 [:,1]
x1=data1 [:,0]
y1=data1 [:,1]
x2=data2 [:,0]
y2=data2 [:,1]
x3=data3 [:,0]
y3=data3 [:,1]

figure, axes = plt.subplots(nrows=2, ncols=2,figsize=(15,15))
axes[0,0].plot(x0,y0)
axes[0,0].scatter(x0,y0,color='r')
axes[0,0].set_xlabel("X")
axes[0,0].set_ylabel("Y")
axes[0,0].set_title("Best Path, Rank 0")

axes[0,1].plot(x1,y1)
axes[0,1].scatter(x1,y1,color='r')
axes[0,1].set_xlabel("X")
axes[0,1].set_ylabel("Y")
axes[0,1].set_title("Best Path, Rank 1")

axes[1,0].plot(x2,y2)
axes[1,0].scatter(x2,y2,color='r')
axes[1,0].set_xlabel("X")
axes[1,0].set_ylabel("Y")
axes[1,0].set_title("Best Path, Rank 2")

axes[1,1].plot(x3,y3)
axes[1,1].scatter(x3,y3,color='r')
axes[1,1].set_xlabel("X")
axes[1,1].set_ylabel("Y")
axes[1,1].set_title("Best Path, Rank 3")
plt.show() 
```

```python
data_file2=np.loadtxt('10.2/dataL0.dat')
data_file3=np.loadtxt('10.2/dataL1.dat')
data_file4=np.loadtxt('10.2/dataL2.dat')
data_file5=np.loadtxt('10.2/dataL3.dat')

Lmin=data_file2 [:,0]
Lmean=data_file2 [:,1]
LdevS=data_file2[:,2]

Lmin1=data_file3 [:,0]
Lmean1=data_file3 [:,1]
LdevS1=data_file3[:,2]

Lmin2=data_file4 [:,0]
Lmean2=data_file4 [:,1]
LdevS2=data_file4 [:,2]

Lmin3=data_file5 [:,0]
Lmean3=data_file5 [:,1]
LdevS3=data_file5[:,2]

figure, axes = plt.subplots(nrows=4, ncols=2,figsize=(15,30))
x2 = np.arange(10001) 
axes[0,0].plot(Lmin)
axes[0,0].set_xlabel("$N_{generations}$")
axes[0,0].set_ylabel("$L_{min}$")
axes[0,0].set_title("$L_{min}$, Rank 0")
axes[0,1].errorbar(x2,Lmean,yerr=LdevS)
axes[0,1].set_xlabel("$N_{generations}$")
axes[0,1].set_ylabel("$L_{mean}$")
axes[0,1].set_title("$L_{mean}$,Rank 0")

axes[1,0].plot(Lmin1)
axes[1,0].set_xlabel("$N_{generations}$")
axes[1,0].set_ylabel("$L_{min}$")
axes[1,0].set_title("$L_{min}$, Rank 1")
axes[1,1].errorbar(x2,Lmean1,yerr=LdevS1)
axes[1,1].set_xlabel("$N_{generations}$")
axes[1,1].set_ylabel("$L_{mean}$")
axes[1,1].set_title("$L_{mean}$,Rank 1")

axes[2,0].plot(Lmin2)
axes[2,0].set_xlabel("$N_{generations}$")
axes[2,0].set_ylabel("$L_{min}$")
axes[2,0].set_title("$L_{min}$, Rank 2")
axes[2,1].errorbar(x2,Lmean2,yerr=LdevS2)
axes[2,1].set_xlabel("$N_{generations}$")
axes[2,1].set_ylabel("$L_{mean}$")
axes[2,1].set_title("$L_{mean}$,Rank 2")

axes[3,0].plot(Lmin2)
axes[3,0].set_xlabel("$N_{generations}$")
axes[3,0].set_ylabel("$L_{min}$")
axes[3,0].set_title("$L_{min}$, Rank 3")
axes[3,1].errorbar(x2,Lmean2,yerr=LdevS2)
axes[3,1].set_xlabel("$N_{generations}$")
axes[3,1].set_ylabel("$L_{mean}$")
axes[3,1].set_title("$L_{mean}$,Rank 3")
plt.show()
```

Anche in questo caso si nota una leggera differenza tra i vari percorsi ma analogamente agli es 09 e 10.1 tale differenza si può imputare alla casualità della misura. Si possono inoltre effettuare le stesse osservazioni che si sono fatte per l'es 09 e 10.1 riguardo alla fluttuazione dei valori di $L_{min}$ e $L_{mean}$.

Si nota comuqnue una coerenza tra le 4 misure anche utilizzando uno scambio di informazione tra i vari nodi.
