# Es 04.1

Si richiede di implementare, nel codice fornitoci per studiare la dinamica molecolare, un metodo che permetta un restart a partire da due configurazioni, e da questo trovare un fattore di riscalamento delle velocità di modo da ottenere un match con la temperatura desiderata.

Per verificare tale ottimizzazione sono partito da una congirazione fcc e ho effettuato 5 run di 10000 step (registrando l'output ogni 10 step) in cui nella prima ho il fattore di scala che dipende da delle variabili random, mentre dalla seconda entra in gioco il nuovo metodo.

I parametri usati per questa ottimizzazione sono tipici di un liquido e sono:
- $T=1.9$
- $N_{particelle}=108$
- $\rho=0.8$
- $r_{cut off}=2.5$
- $\delta t=0.0005$
- $N_{step}=1000$

Dove $\delta t$ è lo step temporale usato nell'algoritmo di Verlet per trovare le configurazioni evolute.

Si mostra sotto il grafico ottenuto a seguito di 5 run diverse per quanto riguarda la T (in azzurro) dove ho mostrato in rosso il valore desiderato della T, le righe verdi verticali indicano la fine di ogni run e il relativo restart.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data_file=np.loadtxt('04.1/Run1_output_temp.dat')
T1=data_file [:]

data_file2=np.loadtxt('04.1/Run2_output_temp.dat')
T2=data_file2 [:]

data_file3=np.loadtxt('04.1/Run3_output_temp.dat')
T3=data_file3 [:]

data_file4=np.loadtxt('04.1/Run4_output_temp.dat')
T4=data_file4 [:]

data_file5=np.loadtxt('04.1/Run5_output_temp.dat')
T5=data_file5 [:]

TTot=np.concatenate((T1,T2,T3,T4,T5))
plt.plot(TTot)
plt.ylabel("T")
plt.xlabel("#step")
plt.title("T")
plt.axhline(y=1.9, color='r', linestyle='-')
plt.axvline(x=1000, color='g', linestyle='--')
plt.axvline(x=2000, color='g', linestyle='--')
plt.axvline(x=3000, color='g', linestyle='--')
plt.axvline(x=4000, color='g', linestyle='--')
plt.show()
```

Si nota che già dopo 3 run abbiamo una buon match con la $T$ desiderata che va migliorando nella run successiva.

Si mostra ora il grafico ottenuto per l'energia cinetica dove il valore atteso è ottenuto a seguito del teorema di equipartizione:

$$E_{kin}=\frac3 2 T$$

```python
dat_file=np.loadtxt('04.1/Run1_output_ekin.dat')
K1=dat_file [:]

dat_file2=np.loadtxt('04.1/Run2_output_ekin.dat')
K2=dat_file2 [:]

dat_file3=np.loadtxt('04.1/Run3_output_ekin.dat')
K3=dat_file3 [:]

dat_file4=np.loadtxt('04.1/Run4_output_ekin.dat')
K4=dat_file4 [:]

dat_file5=np.loadtxt('04.1/Run5_output_ekin.dat')
K5=dat_file5 [:]

KTot=np.concatenate((K1,K2,K3,K4,K5))
plt.plot(KTot)
plt.ylabel("$E_{kin}$")
plt.xlabel("#step")
plt.title("$E_{kin}$")
plt.axhline(y=3*1.9/2, color='r', linestyle='-')
plt.axvline(x=1000, color='g', linestyle='--')
plt.axvline(x=2000, color='g', linestyle='--')
plt.axvline(x=3000, color='g', linestyle='--')
plt.axvline(x=4000, color='g', linestyle='--')
plt.show()
```

Si nota come anche in quaesto caso si osserva un buon match già a partire dalla quarta run.

Tale esperienza è stata effettuata anche per un solido di modo da verificare la correttezza del codice, ma con una sostanziale differenza rispetto alle run precedenti; in questo caso infatti sono stati effettuate 10 run di 1000 step in cui l'output venuva registrato per ogni step. Questa prova l'ho effettuata per verificare se fosse possibile in questo modo velocizzare l'inizializzazione della simulazione.

I parametri usati in questo caso sono:
- $T=0.8$
- $N_{particelle}=108$
- $\rho=1.1$
- $r_{cut off}=2.2$
- $\delta t=0.0005$
- $N_{step}=1000$

```python
data=np.loadtxt("04.1/Es04.1_output_temp.dat")

T=data[:]
plt.plot(T)
plt.ylabel("T")
plt.xlabel("#step")
plt.title("T")
plt.axhline(y=0.8, color='r', linestyle='-')
plt.axvline(x=1000, color='g', linestyle='--')
plt.axvline(x=2000, color='g', linestyle='--')
plt.axvline(x=3000, color='g', linestyle='--')
plt.axvline(x=4000, color='g', linestyle='--')
plt.axvline(x=5000, color='g', linestyle='--')
plt.axvline(x=6000, color='g', linestyle='--')
plt.axvline(x=7000, color='g', linestyle='--')
plt.axvline(x=8000, color='g', linestyle='--')
plt.axvline(x=9000, color='g', linestyle='--')
plt.show()
```

Anche in questo caso si nota un match con la temperatura desiderata dopo 4/5 run anche se i dati sembrano fluttuare attorno al valore atteso molto di più rispetto a prima. Questo però sono diposto ad accettarlo a fronte di un netto miglioramento della prestazione temporale di simulazione.

# Es 04.2

Si richiede di implementare, nel codice fornitoci, un modo di calcolare i valori medi delle grandezze prese in esame. Per fare ciò ho registrato i valori istantanei in un array su cui ho fatto ottenuto i valori medi utilizzando il blocking method.

Ho effettuato la simulazione coi parametri tipici del liquido:
- $T=1.9$
- $N_{particelle}=108$
- $\rho=0.8$
- $r_{cut off}=2.5$
- $\delta t=0.0005$
- $N_{step}=10000$
- $N_{blocks}=1000$

Si mostrano i grafici ottenuti a seguito di ottimizzazione (effettuata nell'esercizio 04.1) con relative barre di errore.

```python
x=np.arange(1000)
x*=10

data_file=np.loadtxt('04.2/2_ave_ekin.out')
K=data_file [:,0]
errK=data_file[:,1]

data_file2=np.loadtxt('04.2/2_ave_epot.out')
U=data_file2 [:,0]
errU=data_file2[:,1]

data_file3=np.loadtxt('04.2/2_ave_etot.out')
E=data_file3 [:,0]
errE=data_file3 [:,1]

data_file4=np.loadtxt('04.2/2_ave_temp.out')
T=data_file4 [:,0]
errT=data_file4[:,1]

figure, axes = plt.subplots(nrows=2, ncols=2,figsize=(15,15))
axes[0,0].errorbar(x,T,yerr=errT)
axes[0,0].set_title('T')
axes[0,0].set_ylabel('T')

axes[0,1].errorbar(x,K,yerr=errK)
axes[0,1].set_title('$E_{kin}$')
axes[0,1].set_ylabel('$E_{kin}$')

axes[1,0].errorbar(x,U,yerr=errU)
axes[1,0].set_title('$E_{pot}$')
axes[1,0].set_ylabel('$E_{pot}$')

axes[1,1].errorbar(x,E,yerr=errE)
axes[1,1].set_title('$E_{tot}$')
axes[1,1].set_ylabel('$E_{pot}$')

for ax in axes.flat:
    ax.set_xlabel('#thorws')

plt.show()
```

In questi grafici non aggiungo la linea orizzontale per il valore atteso (per $T$ e $E_{kin}$) in quanto ho un mismatch significativo dovuto alla casualità del processo che non fanno apprezzare l'andamento del grafico.

Per completezza si mostrano i grafici di energia cinetica, potenziale e totale nello stesso di modo da osservare come sulla scala delle energie l'energia totale risulti costante.

```python
plt.errorbar(x,K,yerr=errK,label='$E_{kin}$')
plt.errorbar(x,U,yerr=errU,label='$E_{pot}$')
plt.errorbar(x,E,yerr=errE,label='$E_{tot}$')
plt.ylabel("$E$")
plt.xlabel("#thorws")
plt.legend()
plt.show()
```

# Es 04.3

Si richiede di effettuare la simulazione di dinamica molecolare per un modello di Argon per le 3 fasi. I parametri per implementare tali esperimenti sono:

1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

Si riportano i grafici ottenuti con le grandezze espresse secondo il SI.

```python
x=np.arange(5000)
x*=20

data_fileS=np.loadtxt('04.3/Solid_ave_ekin.out')
KS=data_fileS [:,0]
errKS=data_fileS[:,1]

data_file2S=np.loadtxt('04.3/Solid_ave_epot.out')
US=data_file2S [:,0]
errUS=data_file2S[:,1]

data_file3S=np.loadtxt('04.3/Solid_ave_etot.out')
ES=data_file3S [:,0]
errES=data_file3S [:,1]

data_file4S=np.loadtxt('04.3/Solid_ave_temp.out')
TS=data_file4S [:,0]
errTS=data_file4S[:,1]

esukb=120
sig=0.34*10**-9
m=39.948*1,66054*10**-27
e=esukb*1.380649*10**-23

TS*=esukb
errTS*=esukb
KS*=e
errKS*=e
US*=e
errUS*=e
ES*=e
errES*=e

figure, axes = plt.subplots(nrows=2, ncols=2,figsize=(15,15))
axes[0,0].errorbar(x,TS,yerr=errTS)
axes[0,0].set_title('T Solid Phase')
axes[0,0].set_ylabel('T(K)')
axes[0,0].axhline(y=0.8*esukb,color='r')

axes[0,1].errorbar(x,KS,yerr=errKS)
axes[0,1].set_title('$E_{kin}$ Solid phase')
axes[0,1].set_ylabel('$E_{kin}$(J)')
axes[0,1].axhline(y=3/2*0.8*e,color='r')

axes[1,0].errorbar(x,US,yerr=errUS)
axes[1,0].set_title('$E_{pot}$ Solid phase')
axes[1,0].set_ylabel('$E_{pot}$(J)')

axes[1,1].errorbar(x,ES,yerr=errES)
axes[1,1].set_title('$E_{tot}$ Solid phase')
axes[1,1].set_ylabel('$E_{tot}$(J)')

for ax in axes.flat:
    ax.set_xlabel('#thorws')

plt.show()
```

```python
data_fileL=np.loadtxt('04.3/Liquid_ave_ekin.out')
KL=data_fileL [:,0]
errKL=data_fileL[:,1]

data_file2L=np.loadtxt('04.3/Liquid_ave_epot.out')
UL=data_file2L [:,0]
errUL=data_file2L[:,1]

data_file3L=np.loadtxt('04.3/Liquid_ave_etot.out')
EL=data_file3L [:,0]
errEL=data_file3L [:,1]

data_file4L=np.loadtxt('04.3/Liquid_ave_temp.out')
TL=data_file4L [:,0]
errTL=data_file4L[:,1]

TL*=esukb
errTL*=esukb
KL*=e
errKL*=e
UL*=e
errUL*=e
EL*=e
errEL*=e

figure, axes = plt.subplots(nrows=2, ncols=2,figsize=(15,15))
axes[0,0].errorbar(x,TL,yerr=errTL)
axes[0,0].set_title('T Liquid Phase')
axes[0,0].set_ylabel('T(K)')
axes[0,0].axhline(y=1.1*esukb,color='r')

axes[0,1].errorbar(x,KL,yerr=errKL)
axes[0,1].set_title('$E_{kin}$ Liquid phase')
axes[0,1].set_ylabel('$E_{kin}$(J)')
axes[0,1].axhline(y=3/2*1.1*e,color='r')

axes[1,0].errorbar(x,UL,yerr=errUL)
axes[1,0].set_title('$E_{pot}$ Liquid phase')
axes[1,0].set_ylabel('$E_{pot}$(J)')

axes[1,1].errorbar(x,EL,yerr=errEL)
axes[1,1].set_title('$E_{tot}$ Liquid phase')
axes[1,1].set_ylabel('$E_{tot}$(J)')

for ax in axes.flat:
    ax.set_xlabel('#thorws')

plt.show()
```

```python
data_fileG=np.loadtxt('04.3/Gas_ave_ekin.out')
KG=data_fileG [:,0]
errKG=data_fileG[:,1]

data_file2G=np.loadtxt('04.3/Gas_ave_epot.out')
UG=data_file2G [:,0]
errUG=data_file2G[:,1]

data_file3G=np.loadtxt('04.3/Gas_ave_etot.out')
EG=data_file3G [:,0]
errEG=data_file3G [:,1]

data_file4G=np.loadtxt('04.3/Gas_ave_temp.out')
TG=data_file4G [:,0]
errTG=data_file4G[:,1]

TG*=esukb
errTG*=esukb
KG*=e
errKG*=e
UG*=e
errUG*=e
EG*=e
errEG*=e

figure, axes = plt.subplots(nrows=2, ncols=2,figsize=(15,15))
axes[0,0].errorbar(x,TG,yerr=errTG)
axes[0,0].set_title('T Gas Phase')
axes[0,0].set_ylabel('T(K)')
axes[0,0].axhline(y=1.2*esukb,color='r')

axes[0,1].errorbar(x,KG,yerr=errKG)
axes[0,1].set_title('$E_{kin}$ Gas phase')
axes[0,1].set_ylabel('$E_{kin}$(J)')
axes[0,1].axhline(y=3/2*1.2*e,color='r')

axes[1,0].errorbar(x,UG,yerr=errUG)
axes[1,0].set_title('$E_{pot}$ Gas phase')
axes[1,0].set_ylabel('$E_{pot}$(J)')

axes[1,1].errorbar(x,EG,yerr=errEG)
axes[1,1].set_title('$E_{tot}$ Gas phase')
axes[1,1].set_ylabel('$E_{tot}$(J)')

for ax in axes.flat:
    ax.set_xlabel('#thorws')

plt.show()
```

Si nota che in tutti e 3 i casi come l'energia totale sia pressochè costante, nella scala delle energie, in tutti e tre i casi in esame. Inoltre si nota un leggero mismatch rispetto alla $T$ e alla $E_{kin}$ attesa, ciò può essere imputabile alla casualità della simulazione.
